### –®–∞–≥ 2.6. –ö–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏—è –¥–æ—Ö–æ–¥–∞.

````python
# –°–æ–∑–¥–∞—ë–º —Ñ—É–Ω–∫—Ü–∏—é —Å —É—Å–ª–æ–≤–∏—è–º–∏
def income_status(total_income):
    if total_income <= 30000:
            return 'E'
    if total_income <= 50000:
            return 'D'
    if total_income <= 200000:
            return 'C'
    if total_income <= 1000000:
            return 'B'
    if total_income >= 1000001:
            return 'A'
````

````python
df['income_status'] = df['total_income'].apply(income_status)
````

````python
# # –ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –±–æ–ª–µ–µ —Ç–æ—á–Ω–æ–≥–æ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è —Ü–µ–ª–µ–π –∫—Ä–µ–¥–∏—Ç–∞
def credit_purpose_status(purpose):
    if '–æ–±—Ä–∞–∑–æ–≤–∞–Ω' in purpose:
            return '–ø–æ–ª—É—á–µ–Ω–∏–µ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è'
    if '–∞–≤—Ç–æ' in purpose:
            return '–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –∞–≤—Ç–æ–º–æ–±–∏–ª–µ–º'
    if '—Å–≤–∞–¥—å' in purpose:
            return '–ø—Ä–æ–≤–µ–¥–µ–Ω–∏–µ —Å–≤–∞–¥—å–±—ã'
    if '–∂–∏–ª—å' or '–Ω–µ–¥–≤' in purpose:
            return '–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å—é'
    return '–ù–µ —É–¥–∞–ª–æ—Å—å –∏–¥–µ–Ω—Ç–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞—Ç—å'
````
